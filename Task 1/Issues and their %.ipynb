```python
import pandas as pd

# Read the CSV file
df = pd.read_csv("review_analyzed.csv")

# Display the dataframe
print(df)
```

```python
# Get the column name with the maximum value for each row
max_topic = df[['Topic 0', 'Topic 1', 'Topic 2', 'Topic 3', 'Topic 4']].idxmax(axis=1)

# Extract the numeric part from the column names
issue = max_topic.str.extract(r'(\d+)').astype(int)

# Assign the extracted values to a new column named 'issue'
df['issue'] = issue

# Display the updated DataFrame
print(df)
```

```python
import matplotlib.pyplot as plt

# Count the occurrences of each issue
issue_counts = df['issue'].value_counts()

# Create a bar chart
plt.bar(issue_counts.index, issue_counts.values)

# Set the x-axis labels
plt.xticks(issue_counts.index)

# Set the y-axis label
plt.ylabel('Count')

# Set the title of the chart
plt.title('Distribution of Issues')

# Display the chart
plt.show()
```

```python
# Select the rows and columns to calculate the sum
sum_topics = df.loc[0:988, 'Topic 0':'Topic 4'].sum()

# Calculate the total sum of topics
total_sum = sum_topics.sum()

# Calculate the percentage of each topic
topic_percentages = (sum_topics / total_sum) * 100

# Create a bar chart
plt.bar(topic_percentages.index, topic_percentages.values)

# Set the x-axis labels
plt.xticks(topic_percentages.index)

# Set the y-axis label
plt.ylabel('Percentage')

# Set the title of the chart
plt.title('Percentage of Prominent Topics')

# Display the chart
plt.show()

# Print the topic percentages
print(topic_percentages)
```
